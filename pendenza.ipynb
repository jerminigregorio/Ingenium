Importazione delle librerie necessarie

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return(x**3+x**2-5*x+2)
```

```python
f(0)
```

Genero il vettore degli argomenti $x$

```python
x=np.linspace(-10,10,21)
print(x)
```

Genero il vettore delle immagini y

```python
y=f(x)
print(y)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Grafico di y=f(x) per x da -10 a 10 e y da -10 a 10')
plt.grid(True)
plt.show()
```

```python
# Genero un vettore x con maggiore dettaglio
x_dettaglio = np.linspace(-10, 10, 201)
y_dettaglio = f(x_dettaglio)

plt.figure(figsize=(8, 6))
plt.plot(x_dettaglio, y_dettaglio)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Grafico dettagliato di y=f(x) per x da -10 a 10 e y da -10 a 10')
plt.grid(True)
plt.show()
```
